8. Создайте случайную матрицу 10х2 которая представляет собой декартовые координаты и преобразуйте их в полярные

````python
#Создаём матрицу
DecartCoordinations = np.random.rand(10, 2)

#Разбиваем по осям
x, y = DecartCoordinations[:, 0], DecartCoordinations[:, 1]

#Посчитаем r и phi для перехода в полярные координаты
r = np.sqrt(pow(x, 2) + pow(y, 2))
phi = np.arctan2(y, x)

#Преобразуем и выводим
PolarCoordinations = np.column_stack((r, phi))
print(PolarCoordinations)
````

````python
import pandas as pd
````

````python
#DataSet лежит в той же директории
DataSet = pd.read_csv('student_perfomance.csv')
````

1. Посчитайте среднее кол-во часов уделяемое на сон студентами

````python
#Применим метод mean к колонке с данными о часах сна и сохраним в отдельную переменную
Mean_Sleep_Hours = DataSet['Sleep_Hours'].mean()
print(Mean_Sleep_Hours)
````

2. Сколько процентов студентов посещают внеклассовые активности?

````python
#В колонке с активностями лежат значения Yes или No. То есть нам нужно посчитать количество Yes и поделить на все, а затем перевести в проценты.
Extracurricular_Percentage = (DataSet['Extracurricular_Activities'] == 'Yes').mean() * 100
print(Extracurricular_Percentage)
````

3. Правда ли, что в среднем девушки сдают экзамены лучше, чем мужчины

````python
#Считаем два средних, в зависимости от пола, а после сравниваем значения. Bool значение спокойно выводится через print()
Mean_Female_Score = DataSet[DataSet['Gender'] == 'female']['Exam_Score'].mean()
Mean_Male_Score = DataSet[DataSet['Gender'] == 'male']['Exam_Score'].mean()
print(Mean_Female_Score > Mean_Male_Score)
````

4. Выведите 25 и 75 перцентили времени сна студентов

````python
#Просто используем встроенную функцию quantile
Sleep_25_Percentile = DataSet['Sleep_Hours'].quantile(0.25)
Sleep_75_Percentile = DataSet['Sleep_Hours'].quantile(0.75)
print(Sleep_25_Percentile, Sleep_75_Percentile)
````

5. Правда ли, что те, кто спит больше(75 перцентиль и выше) сдает экзамен лучше, чем студенты с недосыпом(25 перцентиль и меньше)

````python
#Сравниваем средние, сортируя данные в колонке Sleep_Hours, по посчитанным ранее Sleep_25_Percentile и Sleep_75_Percentile
Worse_Sleepers = DataSet[DataSet['Sleep_Hours'] <= Sleep_25_Percentile]['Exam_Score'].mean()
Better_Sleepers = DataSet[DataSet['Sleep_Hours'] >= Sleep_75_Percentile]['Exam_Score'].mean()
print(Better_Sleepers > Worse_Sleepers)
````

6. Влияет ли физическая активность на результаты экзамена?

````python
#Аналогично выбираем по активности студентов и считаем их средние относительно результатов экзамена. Так как колонка Physical_Activity содержит значения [0;6], то неактивных я считаю как [0;3], а активных как [4;6]. 
Physically_Active_Students = DataSet[DataSet['Physical_Activity'] <= 3]['Exam_Score'].mean()
Physically_Inactive_Students = DataSet[DataSet['Physical_Activity'] > 3]['Exam_Score'].mean()
print(Physically_Active_Students > Physically_Inactive_Students)
````

7. Правда ли, что слуденты, чьи родители имеют высшее образование, сдают экзамены лучше, чем остальные?

````python
#В колонке Parental_Education_Level только 3 значения: College, High School и Postgraduate. Высшее образование имеют только Postgraduate (аспирантура), так как в РФ College не считается высшим, а High School - 10 и 11 класс.
Educated_Parents = DataSet[DataSet['Parental_Education_Level'] == 'Postgraduate']['Exam_Score'].mean()
Other_Parents = DataSet[DataSet['Parental_Education_Level'] != 'Postgraduate']['Exam_Score'].mean()
print(Educated_Parents > Other_Parents)
````

8. Как влияет участие родителей на результаты экзаменов

````python
 #По сути нас просят оценить корреляцию данных. Для этого используем функцию corr. Однако значения Parental_Involvement 'Low', 'Medium', 'High', то есть строки. Значит, нужно использовать коды.
DataSet['Parental_Involvement_Code'] = pd.Categorical(DataSet['Parental_Involvement']).codes

#Значения corr [-1;1]. Чем ближе к 1, тем больше корреляция данных и наоборот.
Parental_Involvement_Correlation = DataSet['Parental_Involvement_Code'].corr(DataSet['Exam_Score'])
print(Parental_Involvement_Correlation)
````

9. Правда ли, что в семьях с доходом больше среднего дети сдают экзамены лучше?

````python
#Так как значения лежат строковые - кодируем.
DataSet['Family_Income_Code'] = pd.Categorical(DataSet['Family_Income']).codes

#Считаем среднее значение
Mean_Income = DataSet['Family_Income_Code'].mean()

#Сортируем на две категории: выше среднего и ниже среднего.
Higher_Income_Students = DataSet[DataSet['Family_Income_Code'] > Mean_Income]['Exam_Score'].mean()
Lower_Income_Students = DataSet[DataSet['Family_Income_Code'] <= Mean_Income]['Exam_Score'].mean()

print(Higher_Income_Students > Lower_Income_Students)
````

10. Посчитайте долю студентов сдавших экзамен на более чем 70 баллов в семьях с доходом ниже 25го перцентиля и отсутствием доступа к интернету

````python
#Доход ниже 25 перцентиля
Income_25_Percentile = DataSet['Family_Income_Code'].quantile(0.25)

#Сортируем по заданным условиям.
Low_Income_No_Internet = DataSet[(DataSet['Family_Income_Code'] <= Income_25_Percentile) & (DataSet['Internet_Access'] == 'No')]
High_Score_Low_Income_Students = Low_Income_No_Internet[Low_Income_No_Internet['Exam_Score'] > 70]
High_Score_Students = DataSet['Exam_Score'] > 70

#В конечном итоге считаем долю, сколько среди высокобальников студентов в семьях с доходом ниже 25го перцентиля и отсутствием доступа к интернету.
Percentage_Of_High_Score = len(High_Score_Low_Income_Students) / len(High_Score_Students)
print(Percentage_Of_High_Score)
````
